```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import numpy as np

import brewer2mpl
# brewer2mpl.get_map args: set name set type number of colors
bmap = brewer2mpl.get_map('Pastel2', 'qualitative', 3)
new_colors = bmap.mpl_colors


colors = ['lightcoral', 'orange', 'lightgreen', 'cornflowerblue', 'darkviolet', 'seashell', 'honeydew', 'lightcyan']
markers = ['*', 'd', 's', '^', 'o']
font_size = 35
markersize = 20
plt.figure(figsize=(60, 12))            # 
plt.rcParams['font.sans-serif'] = ['Nimbus Roman']  # 
x_label = 'Bandwidth of edge node (MHz)'
#################################################################### #################################################################
plt.subplot(141)                        # 

plt.xlabel(x_label, fontsize=font_size)                          # xlabel
plt.ylabel('Cumulative reward', fontsize=font_size)          # ylabel     plotfontsize
plt.title('(a) CR', fontsize=font_size, y=-0.2)     # title

x_vals = [1, 2, 3, 4, 5]
ra_ra = [175.06426381203300, 214.060371670128, 240.467731991647, 263.04380146898, 281.303935240277]
c_ddpg_ra = [208.748886974234, 259.291964896801, 307.999658582648, 322.878081457575, 354.6059242451440]
iac_ra = [248.59473010595800, 269.830509433376, 290.4434458219750, 331.737999886021, 351.356965777912]
iac_ga_ra = [236.87664323697, 318.508683985828, 315.034869116743, 354.157492999339, 373.360464430689]
madr_ga_ra = [261.357364093338, 330.663875264952, 357.196511301275, 361.257326654576, 381.3391307319460]

plt.plot(x_vals, ra_ra, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, c_ddpg_ra, linestyle="-", marker=markers[1], markersize=markersize, color=colors[1], label= "C-DDPG", linewidth=5.0)
plt.plot(x_vals, iac_ra, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "MAC", linewidth=5.0)
plt.plot(x_vals, iac_ga_ra, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "MAC-GBA", linewidth=5.0)
plt.plot(x_vals, madr_ga_ra, linestyle="-", marker=markers[4], markersize=markersize, color=colors[4], label= "MDR-GBA", linewidth=5.0)
plt.grid()
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
plt.ylim(100, 400)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0
#################################################################### #################################################################
plt.subplot(142)

plt.xlabel(x_label, fontsize=font_size)
plt.ylabel('Average reward proportion', fontsize=font_size)
plt.title('(b) ARP', fontsize=font_size, y=-0.2)     # title

x_vals = [1, 2, 3, 4, 5]
ra_timeliness =         [173.261328543694, 139.299703540036, 119.49040101647, 96.6604635740394, 82.48930805370330]
c_ddpg_timeliness =     [127.497419859678, 92.7357002326294, 63.399466618854, 54.8038847269749, 38.8931217024911]
iac_timeliness =        [108.31190279602600, 87.36828877508380, 71.86936523555500, 50.8749500463991, 41.3263206771922]
iac_ga_timeliness =     [119.742958749536, 64.08676921253360, 62.3100290334979, 42.9438940653106, 32.5195750359184]
madr_ga_timeliness =    [104.40967127088000, 59.2325289139396, 44.8817911744236, 38.0536649865231, 30.9645657519657]

ra_consistence =        [1356.02625266665, 1087.61186656748, 916.0638812925400, 781.423701216166, 682.95634267932]
c_ddpg_consistence =    [1165.96300172739, 858.0798608970620, 576.2748987485260, 503.58398646124400, 350.670690844294]
iac_consistence =       [774.2129578472700, 685.860022256558, 628.0975888313910, 421.491732308324, 338.965281332663]
iac_ga_consistence =    [899.2253145576010, 473.678178443449, 506.963974031169, 321.819243511184, 251.125081159668]
madr_ga_consistence =   [769.0818384023810, 398.989823475997, 274.902607644425, 305.293314473337, 197.996298986449]

ra_completeness =       [0.54802658643714, 0.6374684777850690, 0.694646023590775, 0.739525500772115, 0.772347901377529]
c_ddpg_completeness =   [0.6113456660908690, 0.7139733797009800, 0.8079083670838250, 0.8321386711795860, 0.88310976971857]
iac_completeness =      [0.7419290140509110, 0.7713799925811480, 0.7906341370562040, 0.8595027558972260, 0.887011572889113]
iac_ga_completeness =   [0.7002582284808000, 0.842107273852184, 0.8310120086562770, 0.8927269188296060, 0.9162916396134440]
madr_ga_completeness =  [0.7436393871992070, 0.8670033921746680, 0.908365797451858, 0.9097624190457150, 0.934001233671184]

for i in range(len(ra_timeliness)):
    ra_timeliness[i] = 3 / 10 * (1 - np.tanh(ra_timeliness[i] / 100))
    c_ddpg_timeliness[i] = 3 / 10 * (1 - np.tanh(c_ddpg_timeliness[i] / 100))
    iac_timeliness[i] = 3 / 10 * (1 - np.tanh(iac_timeliness[i] / 100))
    iac_ga_timeliness[i] = 3 / 10 * (1 - np.tanh(iac_ga_timeliness[i] / 100))
    madr_ga_timeliness[i] = 3 / 10 * (1 - np.tanh(madr_ga_timeliness[i] / 100))

for i in range(len(ra_consistence)):
    ra_consistence[i] = 3 / 10 * (1 - np.tanh(ra_consistence[i] / 1000))
    c_ddpg_consistence[i] = 3 / 10 * (1 - np.tanh(c_ddpg_consistence[i] / 1000))
    iac_consistence[i] = 3 / 10 * (1 - np.tanh(iac_consistence[i] / 1000))
    iac_ga_consistence[i] = 3 / 10 * (1 - np.tanh(iac_ga_consistence[i] / 1000))
    madr_ga_consistence[i] = 3 / 10 * (1 - np.tanh(madr_ga_consistence[i] / 1000))

for i in range(len(ra_completeness)):
    ra_completeness[i] = 4 / 10 * ra_completeness[i]
    c_ddpg_completeness[i] = 4 / 10 * c_ddpg_completeness[i]
    iac_completeness[i] = 4 / 10 * iac_completeness[i]
    iac_ga_completeness[i] = 4 / 10 * iac_ga_completeness[i]
    madr_ga_completeness[i] = 4 / 10 * madr_ga_completeness[i]

diff = 1
x = np.arange(1, 6, 1)

width = diff * 0.12
small_width = diff * 0.03
label_text_size = 30

ra_cp = plt.bar(x - width - 8 * small_width, ra_completeness, width, label='Completeness', hatch="/", color=new_colors[0], edgecolor='k')
ra_cs = plt.bar(x - width - 8 * small_width, ra_consistence, width, bottom=ra_completeness, label='Consistency', hatch="\\", color=new_colors[1], edgecolor='k')
ra_tl = plt.bar(x - width - 8 * small_width, ra_timeliness, width, bottom=[ra_completeness[i]+ra_consistence[i] for i in range(len(ra_completeness))], label='Timeliness', hatch="X", color=new_colors[2], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - width - 11 * small_width, ra_completeness[i]+ra_consistence[i]+ra_timeliness[i]+0.03, 'RA', fontsize=label_text_size, rotation=90)

c_ddpg_cp = plt.bar(x - width - 2 * small_width, c_ddpg_completeness, width, hatch="/", color=new_colors[0], edgecolor='k')
c_ddpg_cs = plt.bar(x - width - 2 * small_width, c_ddpg_consistence, width, bottom=c_ddpg_completeness, hatch="\\", color=new_colors[1], edgecolor='k')
c_ddpg_tl = plt.bar(x - width - 2 * small_width, c_ddpg_timeliness, width, bottom=[c_ddpg_completeness[i]+c_ddpg_consistence[i] for i in range(len(c_ddpg_completeness))], hatch="X", color=new_colors[2], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - width - 5 * small_width, c_ddpg_completeness[i]+c_ddpg_consistence[i]+c_ddpg_timeliness[i]+0.03, 'C-DDPG', fontsize=label_text_size, rotation=90)

iac_cp = plt.bar(x - width + 4 * small_width, iac_completeness, width, hatch="/", color=new_colors[0], edgecolor='k')
iac_cs = plt.bar(x - width + 4 * small_width, iac_consistence, width, bottom=iac_completeness, hatch="\\", color=new_colors[1], edgecolor='k')
iac_tl = plt.bar(x - width + 4 * small_width, iac_timeliness, width, bottom=[iac_completeness[i]+iac_consistence[i] for i in range(len(iac_completeness))], hatch="X", color=new_colors[2], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - width + small_width, iac_completeness[i]+iac_consistence[i]+iac_timeliness[i]+0.03, 'MAC', fontsize=label_text_size, rotation=90)

iac_ga_cp = plt.bar(x + 6 * small_width, iac_ga_completeness, width, hatch="/", color=new_colors[0], edgecolor='k')
iac_ga_cs = plt.bar(x + 6 * small_width, iac_ga_consistence, width, bottom=iac_ga_completeness, hatch="\\", color=new_colors[1], edgecolor='k')
iac_ga_tl = plt.bar(x + 6 * small_width, iac_ga_timeliness, width, bottom=[iac_ga_completeness[i]+iac_ga_consistence[i] for i in range(len(iac_ga_completeness))], hatch="X", color=new_colors[2], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] + 3 * small_width, iac_ga_completeness[i]+iac_ga_consistence[i]+iac_ga_timeliness[i]+0.03, 'MAC-GBA', fontsize=label_text_size, rotation=90)

madr_ga_cp = plt.bar(x + width + 8 * small_width, madr_ga_completeness, width, hatch="/", color=new_colors[0], edgecolor='k')
madr_ga_cs = plt.bar(x + width + 8 * small_width, madr_ga_consistence, width, bottom=madr_ga_completeness, hatch="\\", color=new_colors[1], edgecolor='k')
madr_ga_tl = plt.bar(x + width + 8 * small_width, madr_ga_timeliness, width, bottom=[madr_ga_completeness[i]+madr_ga_consistence[i] for i in range(len(madr_ga_completeness))], hatch="X", color=new_colors[2], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] + width + 5 * small_width, madr_ga_completeness[i]+madr_ga_consistence[i]+madr_ga_timeliness[i]+0.03, 'MDR-GBA', fontsize=label_text_size, rotation=90)

plt.grid()

plt.ylim(0, 1.2)  
# 
plt.tick_params(labelsize=font_size)
plt.legend(fontsize = label_text_size, loc=2)
#################################################################### #################################################################
plt.subplot(143)                  # 

plt.xlabel(x_label, fontsize=font_size)                          # xlabel
plt.ylabel('Average queuing time', fontsize=font_size)          # ylabel     plotfontsize
plt.title('(c) AQT', fontsize=font_size, y=-0.2)     # title

x_vals = [1, 2, 3, 4, 5]
ra_st =        [0.0435425893414394, 0.0375305184410894, 0.0330100689889741, 0.0292687350897314, 0.0259262015330401]
c_ddpg_st =    [0.0334809933227276, 0.0313740474592653, 0.0260876706945195, 0.0292354583254089, 0.028122273510779100]
iac_st =       [0.02807764868670920, 0.0253160023451373, 0.028804853790378, 0.0256688748350439, 0.023409927977632]
iac_ga_st =    [0.0389224837550549, 0.0251340918267102, 0.0258922627511766, 0.024937134824880600, 0.0249223258803499]
madr_ga_st =   [0.023789228054564, 0.025302906935621, 0.024692191782083700, 0.025136174353790500, 0.0227857326309102]

plt.plot(x_vals, ra_st, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, c_ddpg_st, linestyle="-", marker=markers[1], markersize=markersize, color=colors[1], label= "C-DDPG", linewidth=5.0)
plt.plot(x_vals, iac_st, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "MAC", linewidth=5.0)
plt.plot(x_vals, iac_ga_st, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "MAC-GBA", linewidth=5.0)
plt.plot(x_vals, madr_ga_st, linestyle="-", marker=markers[4], markersize=markersize, color=colors[4], label= "MDR-GBA", linewidth=5.0)

plt.grid()
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))

plt.ylim(0.02, 0.05)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0

#################################################################### #################################################################
plt.subplot(144)                        # 

plt.xlabel(x_label, fontsize=font_size)                          # xlabel
plt.ylabel('Service rate', fontsize=font_size)          # ylabel     plotfontsize
plt.title('(d) SR', fontsize=font_size, y=-0.2)     # title

x_vals = [1, 2, 3, 4, 5]
ra_sr =         [0.549716202270382, 0.639265565875473, 0.696403696097164, 0.740866873065015, 0.773477812177503]
c_ddpg_sr =     [0.614193326453388, 0.716152884638309, 0.810001123951405, 0.83390092879257, 0.884313044307527]
iac_sr =        [0.740798073615411,	0.771565788638739, 0.785779971185974, 0.856511867905057, 0.883135534915721]
iac_ga_sr =     [0.701306508359814, 0.842328350482107, 0.831678059446811, 0.891899716287418, 0.914266349235544]
madr_ga_sr =    [0.743254213966288, 0.867174062607499, 0.908292053663571, 0.909700722394221, 0.933264533883729]

plt.plot(x_vals, ra_sr, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, c_ddpg_sr, linestyle="-", marker=markers[1], markersize=markersize, color=colors[1], label="C-DDPG", linewidth=5.0)
plt.plot(x_vals, iac_sr, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label="MAC", linewidth=5.0)
plt.plot(x_vals, iac_ga_sr, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "MAC-GBA", linewidth=5.0)
plt.plot(x_vals, madr_ga_sr, linestyle="-", marker=markers[4], markersize=markersize, color=colors[4], label= "MDR-GBA", linewidth=5.0)

plt.ylim(0.4, 1)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0
ax = plt.gca()
ax.xaxis.set_major_locator(MultipleLocator(1))
plt.grid()
plt.savefig('different_bandwidth.png', dpi=50)
plt.show()
```
