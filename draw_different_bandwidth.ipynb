```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['cornflowerblue', 'lightgreen', 'gold', 'tomato']
markers = ['*', 'x', 's', 'o']
font_size = 70
markersize = 50
plt.figure(figsize=(160, 30))            # 
#################################################################### #################################################################
plt.subplot(141)                        # 

plt.xlabel('Bandwidth of edge node (MHz)', fontsize=font_size)                          # xlabel
plt.ylabel('Cumulative reward', fontsize=font_size)          # ylabel     plotfontsize
plt.title('CR', fontsize=font_size)     # title

x_vals = [2, 2.5, 3, 3.5, 4]
ra_ra = [53.749409547185, 53.52306298511680, 54.06977369271610, 58.613940294885, 60.6461268675445]
iddpg_ra = [67.57345904898910, 56.5322978923877, 57.729080859467600, 67.94467236987500, 66.6161031864577]
hmaimd_ra = [86.6385814324735, 78.5890357936904, 88.00810805123510, 93.2825122545582, 94.56744413442360]

plt.plot(x_vals, ra_ra, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_ra, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_ra, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)
plt.grid()
# plt.yplt.grid()lim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0
#################################################################### #################################################################
plt.subplot(142)

plt.xlabel('Bandwidth of edge node (MHz)', fontsize=font_size)
plt.ylabel('AoV proportion', fontsize=font_size)
plt.title('AoVP', fontsize=font_size)

x_vals = [2, 2.5, 3, 3.5, 4]
ra_timeliness =         [64.38908695297730, 90.8556826185388, 93.2514041805764, 73.1197741355415, 67.91812241346050]
iddpg_timeliness =      [39.666178052613, 56.5322978923877, 57.729080859467600, 28.1300995282377, 41.42266570315980]
hmaimd_timeliness =     [38.89163090619830, 50.4601295341915, 64.3284317899197, 64.94753940513230, 33.3427938253676]

ra_consistence =        [2744.18527493774, 3183.69215092417, 3130.5700396244900, 2368.10858420798, 2295.43127926195]
iddpg_consistence =      [1244.54913887334, 1676.70816748091, 2703.92761698647, 2947.95686579206, 2359.53701668415]
hmaimd_consistence =    [445.548959779405, 1500.62898282978, 752.39099015685, 396.143780983068, 1174.6255607624]

ra_completeness =       [0.45070172655957, 0.453785038944353, 0.44841316838345, 0.47404153349374, 0.48343547088351]
iddpg_completeness =    [0.456327381451156, 0.403527585450135, 0.416158396877269, 0.469989346963857, 0.469267629601698]
hmaimd_completeness =   [0.564309221152849, 0.588951817336621, 0.593597006751419, 0.589455333700431, 0.5850330509269230]

for i in range(len(ra_timeliness)):
    ra_timeliness[i] = 1 / 3 * (1 - np.tanh(ra_timeliness[i] / 100))
    iddpg_timeliness[i] = 1 / 3 * (1 - np.tanh(iddpg_timeliness[i] / 100))
    hmaimd_timeliness[i] = 1 / 3 * (1 - np.tanh(hmaimd_timeliness[i] / 100))
    # ra_timeliness[i] = 3 / 10 * (1 - np.tanh(ra_timeliness[i] / 100))
    # iddpg_timeliness[i] = 3 / 10 * (1 - np.tanh(iddpg_timeliness[i] / 100))
    # hmaimd_timeliness[i] = 3 / 10 * (1 - np.tanh(hmaimd_timeliness[i] / 100))

for i in range(len(ra_consistence)):
    # ra_consistence[i] = 3 / 10 * (1 - np.tanh(ra_consistence[i] / 1000))
    # iddpg_consistence[i] = 3 / 10 * (1 - np.tanh(iddpg_consistence[i] / 1000))
    # hmaimd_consistence[i] = 3 / 10 * (1 - np.tanh(hmaimd_consistence[i] / 1000))
    ra_consistence[i] = 1 / 3 * (1 - np.tanh(ra_consistence[i] / 1000))
    iddpg_consistence[i] = 1 / 3 * (1 - np.tanh(iddpg_consistence[i] / 1000))
    hmaimd_consistence[i] = 1 / 3 * (1 - np.tanh(hmaimd_consistence[i] / 1000))

for i in range(len(ra_completeness)):
    # ra_completeness[i] = 4 / 10 * ra_completeness[i]
    # iddpg_completeness[i] = 4 / 10 * iddpg_completeness[i]
    # hmaimd_completeness[i] = 4 / 10 * hmaimd_completeness[i]
    ra_completeness[i] = 1 / 3 * ra_completeness[i]
    iddpg_completeness[i] = 1 / 3 * iddpg_completeness[i]
    hmaimd_completeness[i] = 1 / 3 * hmaimd_completeness[i]

x = np.arange(2, 4.5, 0.5)

width = 0.1

ra_cp = plt.bar(x - width + 0.01, ra_completeness, width, label='Completeness', hatch=".", color=colors[1], edgecolor='k')
ra_cs = plt.bar(x - width + 0.01, ra_consistence, width, bottom=ra_completeness, label='Consistence', hatch="-", color=colors[2], edgecolor='k')
ra_tl = plt.bar(x - width + 0.01, ra_timeliness, width, bottom=[ra_completeness[i]+ra_consistence[i] for i in range(len(ra_completeness))], label='Timeliness', hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - width + 0.01 - 0.05, ra_completeness[i]+ra_consistence[i]+ra_timeliness[i]+0.02, 'RA', fontsize=font_size, rotation=90)

iddpg_cp = plt.bar(x + 0.02, iddpg_completeness, width, hatch=".", color=colors[1], edgecolor='k')
iddpg_cs = plt.bar(x + 0.02, iddpg_consistence, width, bottom=iddpg_completeness, hatch="-", color=colors[2], edgecolor='k')
iddpg_tl = plt.bar(x + 0.02, iddpg_timeliness, width, bottom=[iddpg_completeness[i]+iddpg_consistence[i] for i in range(len(iddpg_completeness))], hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] + 0.02  - 0.05, iddpg_completeness[i]+iddpg_consistence[i]+iddpg_timeliness[i]+0.02, 'IDDPG', fontsize=font_size, rotation=90)

hmaimd_cp = plt.bar(x + width + 0.03, hmaimd_completeness, width, hatch=".", color=colors[1], edgecolor='k')
hmaimd_cs = plt.bar(x + width + 0.03, hmaimd_consistence, width, bottom=hmaimd_completeness, hatch="-", color=colors[2], edgecolor='k')
hmaimd_tl = plt.bar(x + width + 0.03, hmaimd_timeliness, width, bottom=[hmaimd_completeness[i]+hmaimd_consistence[i] for i in range(len(hmaimd_completeness))], hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] + width + 0.03  - 0.05, hmaimd_completeness[i]+hmaimd_consistence[i]+hmaimd_timeliness[i]+0.02, 'HMAIMD', fontsize=font_size, rotation=90)

plt.grid()

plt.ylim(0, 0.8)  
# 
plt.tick_params(labelsize=font_size)
plt.legend(loc=0, fontsize = font_size)
#################################################################### #################################################################
plt.subplot(143)                        # 

plt.xlabel('Bandwidth of edge node (MHz)', fontsize=font_size)                          # xlabel
plt.ylabel('Average service time', fontsize=font_size)          # ylabel     plotfontsize
plt.title('AST', fontsize=font_size)     # title

x_vals = [2, 2.5, 3, 3.5, 4]
ra_st = [0.19946664020734800, 0.191692687267574, 0.499640357323587, 0.178008485279251, 0.289243070334166]
iddpg_st = [0.155920008474464, 0.6319167987712940, 0.19712533222377300, 0.178272750752404, 0.286471410310928]
hmaimd_st = [0.250609755157702, 0.1651995238684160, 0.16292543790714200, 0.13640962613677900, 0.460794743614659]

plt.plot(x_vals, ra_st, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_st, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_st, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)
plt.grid()
# plt.ylim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0

#################################################################### #################################################################
plt.subplot(144)                        # 

plt.xlabel('Bandwidth of edge node (MHz)', fontsize=font_size)                          # xlabel
plt.ylabel('Service rate', fontsize=font_size)          # ylabel     plotfontsize
plt.title('SR', fontsize=font_size)     # title

x_vals = [2, 2.5, 3, 3.5, 4]
ra_sr = [0.107611111111111, 0.118833333333333, 0.0731944444444445, 0.121527777777778, 0.144777777777778]
iddpg_sr = [0.0274166666666667, 0.0413611111111111, 0.0113055555555556, 0.0308611111111111, 0.0881666666666666]
hmaimd_sr = [0.235944444444444, 0.3398611111111110, 0.359944444444445, 0.368222222222223, 0.334777777777778]

plt.plot(x_vals, ra_sr, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_sr, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label="IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_sr, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)

# plt.ylim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0

plt.grid()
plt.savefig('different_bandwidth.png', dpi=50)
plt.show()
```
