```python
import matplotlib.pyplot as plt
import numpy as np

colors = ['cornflowerblue', 'lightgreen', 'gold', 'tomato']
markers = ['*', 'x', 's', 'o']
font_size = 70
markersize = 50
plt.figure(figsize=(160, 30))            # 
#################################################################### #################################################################
plt.subplot(141)                        # 

plt.xlabel('Edge view required by edge node', fontsize=font_size)                          # xlabel
plt.ylabel('Cumulative reward', fontsize=font_size)          # ylabel     plotfontsize
plt.title('CR', fontsize=font_size)     # title

x_vals = [150, 300, 450, 600, 750]
ra_ra = [55.6662121605636, 71.95361551167790, 54.06977369271610, 87.92616415045780, 79.4082382771563]
iddpg_ra = [67.638913337357, 71.85670509129950, 57.729080859467600, 111.37377138656500, 101.02577768306400]
hmaimd_ra = [92.88162911024180, 81.6899262315507, 88.00810805123510, 118.332594926286, 109.89604499570100]

plt.plot(x_vals, ra_ra, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_ra, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_ra, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)
plt.grid()
# plt.yplt.grid()lim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0
#################################################################### #################################################################
plt.subplot(142)

plt.xlabel('Edge view required by edge node', fontsize=font_size)
plt.ylabel('AoV proportion', fontsize=font_size)
plt.title('AoVP', fontsize=font_size)

x_vals = [150, 300, 450, 600, 750]
ra_timeliness =         [79.74914090784290, 87.4056619292562, 93.2514041805764, 63.76160191581110, 74.06032848017590]
iddpg_timeliness =      [72.14034968730290, 73.9200892051441, 57.729080859467600, 26.696521290653, 38.51455800821950]
hmaimd_timeliness =     [80.26115412857410, 73.14530529705930, 64.3284317899197, 22.1522372074438, 33.951123102451900]

ra_consistence =        [1808.60648018259, 2764.72488299366, 3130.5700396244900, 1107.3857632116000, 1951.39427979027]
iddpg_consistence =     [1418.31824811599, 2487.115487231680, 2703.92761698647, 242.06109963761100, 350.0029146835400]
hmaimd_consistence =    [0, 1184.5211818789200, 752.39099015685, 279.269567254917, 178.7827827676070]

ra_completeness =       [0.382275069928256, 0.524190247891179, 0.44841316838345, 0.5429796529723000, 0.52960216534857]
iddpg_completeness =    [0.487801855921856, 0.5304157958055020, 0.416158396877269, 0.5901679366618440, 0.5809283928671180]
hmaimd_completeness =   [0.5284069579113700, 0.5742226525653000, 0.593597006751419, 0.6141701804083460, 0.5875903660609540]

for i in range(len(ra_timeliness)):
    ra_timeliness[i] = 1 / 3 * (1 - np.tanh(ra_timeliness[i] / 100))
    iddpg_timeliness[i] = 1 / 3 * (1 - np.tanh(iddpg_timeliness[i] / 100))
    hmaimd_timeliness[i] = 1 / 3 * (1 - np.tanh(hmaimd_timeliness[i] / 100))
    # ra_timeliness[i] = 3 / 10 * (1 - np.tanh(ra_timeliness[i] / 100))
    # iddpg_timeliness[i] = 3 / 10 * (1 - np.tanh(iddpg_timeliness[i] / 100))
    # hmaimd_timeliness[i] = 3 / 10 * (1 - np.tanh(hmaimd_timeliness[i] / 100))

for i in range(len(ra_consistence)):
    ra_consistence[i] = 1 / 3 * (1 - np.tanh(ra_consistence[i] / 1000))
    iddpg_consistence[i] = 1 / 3 * (1 - np.tanh(iddpg_consistence[i] / 1000))
    hmaimd_consistence[i] = 1 / 3 * (1 - np.tanh(hmaimd_consistence[i] / 1000))
    # ra_consistence[i] = 3 / 10 * (1 - np.tanh(ra_consistence[i] / 1000))
    # iddpg_consistence[i] = 3 / 10 * (1 - np.tanh(iddpg_consistence[i] / 1000))
    # hmaimd_consistence[i] = 3 / 10 * (1 - np.tanh(hmaimd_consistence[i] / 1000))

for i in range(len(ra_completeness)):
    ra_completeness[i] = 1 / 3 * ra_completeness[i]
    iddpg_completeness[i] = 1 / 3 * iddpg_completeness[i]
    hmaimd_completeness[i] = 1 / 3 * hmaimd_completeness[i]
    # ra_completeness[i] = 4 / 10 * ra_completeness[i]
    # iddpg_completeness[i] = 4 / 10 * iddpg_completeness[i]
    # hmaimd_completeness[i] = 4 / 10 * hmaimd_completeness[i]

x = np.arange(150, 900, 150)

width = 25

ra_cp = plt.bar(x - width - 5, ra_completeness, width, label='Completeness', hatch=".", color=colors[1], edgecolor='k')
ra_cs = plt.bar(x - width - 5, ra_consistence, width, bottom=ra_completeness, label='Consistence', hatch="-", color=colors[2], edgecolor='k')
ra_tl = plt.bar(x - width - 5, ra_timeliness, width, bottom=[ra_completeness[i]+ra_consistence[i] for i in range(len(ra_completeness))], label='Timeliness', hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - width - 16, ra_completeness[i]+ra_consistence[i]+ra_timeliness[i]+0.02, 'RA', fontsize=font_size, rotation=90)

iddpg_cp = plt.bar(x , iddpg_completeness, width, hatch=".", color=colors[1], edgecolor='k')
iddpg_cs = plt.bar(x , iddpg_consistence, width, bottom=iddpg_completeness, hatch="-", color=colors[2], edgecolor='k')
iddpg_tl = plt.bar(x , iddpg_timeliness, width, bottom=[iddpg_completeness[i]+iddpg_consistence[i] for i in range(len(iddpg_completeness))], hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] - 11 , iddpg_completeness[i]+iddpg_consistence[i]+iddpg_timeliness[i]+0.02, 'IDDPG', fontsize=font_size, rotation=90)

hmaimd_cp = plt.bar(x + width + 5, hmaimd_completeness, width, hatch=".", color=colors[1], edgecolor='k')
hmaimd_cs = plt.bar(x + width + 5, hmaimd_consistence, width, bottom=hmaimd_completeness, hatch="-", color=colors[2], edgecolor='k')
hmaimd_tl = plt.bar(x + width + 5, hmaimd_timeliness, width, bottom=[hmaimd_completeness[i]+hmaimd_consistence[i] for i in range(len(hmaimd_completeness))], hatch="X", color=colors[3], edgecolor='k')

for i in range(len(x_vals)):
    plt.text(x[i] + width + 3  - 10, hmaimd_completeness[i]+hmaimd_consistence[i]+hmaimd_timeliness[i]+0.02, 'HMAIMD', fontsize=font_size, rotation=90)

plt.grid()

plt.ylim(0, 1)  
# 
plt.tick_params(labelsize=font_size)
plt.legend(loc=0, fontsize = font_size)
#################################################################### #################################################################
plt.subplot(143)                        # 

plt.xlabel('Edge view required by edge node', fontsize=font_size)                          # xlabel
plt.ylabel('Average service time', fontsize=font_size)          # ylabel     plotfontsize
plt.title('AST', fontsize=font_size)     # title

x_vals = [150, 300, 450, 600, 750]
ra_st = [0.19063289907528000, 0.276883606537343, 0.499640357323587, 0.188412209043715, 0.206836799728702]
iddpg_st = [0.195688392747946, 0.2685920022878040, 0.19712533222377300, 0.19515158243906400, 0.19590446750446100]
hmaimd_st = [0.12663700489785100, 0.21736061741285600, 0.16292543790714200, 0.220749534937726, 0.23934057196520600]

plt.plot(x_vals, ra_st, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_st, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label= "IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_st, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)
plt.grid()
# plt.ylim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0

#################################################################### #################################################################
plt.subplot(144)                        # 

plt.xlabel('Edge view required by edge node', fontsize=font_size)                          # xlabel
plt.ylabel('Service rate', fontsize=font_size)          # ylabel     plotfontsize
plt.title('SR', fontsize=font_size)     # title

x_vals = [150, 300, 450, 600, 750]
ra_sr = [0.0325555555555556, 0.2624, 0.0731944444444445, 0.170646825396825, 0.202277777777778]
iddpg_sr = [0.102683333333333, 0.26294444444444400, 0.0113055555555556, 0.3201349206349210, 0.24388888888888900]
hmaimd_sr = [0.260544444444445, 0.349388888888889, 0.359944444444445, 0.376137301587302, 0.29244444444444400]

plt.plot(x_vals, ra_sr, linestyle="-", marker=markers[0], markersize=markersize, color=colors[0], label="RA", linewidth=5.0)
plt.plot(x_vals, iddpg_sr, linestyle="-", marker=markers[2], markersize=markersize, color=colors[2], label="IDDPG", linewidth=5.0)
plt.plot(x_vals, hmaimd_sr, linestyle="-", marker=markers[3], markersize=markersize, color=colors[3], label= "HMAIMD", linewidth=5.0)

# plt.ylim(0, 5)                     # y0,5y05
# 
plt.tick_params(labelsize=font_size)        # 
plt.legend(loc=0, fontsize = font_size)     # loc=0

plt.grid()
plt.savefig('different_edge_view.png', dpi=50)
plt.show()
```
